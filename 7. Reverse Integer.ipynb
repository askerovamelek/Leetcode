```python
class Solution:
    def reverse(self, x: int) -> int: 
        if x > 0:
            a = str(x)
            b = a[::-1]
            m = int(b)
        else:
            a = str(abs(x))
            b = a[::-1]
            m = -int(b)

        # 32-bit limit yoxlamasÄ±
        if m < -2**31 or m > 2**31 - 1:
            return 0

        return m
```
