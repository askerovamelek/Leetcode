```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        left=0
        only_letters = "".join(ch for ch in s if ch.isalnum())
        only_letters=only_letters.lower()
        right=len(only_letters)-1
        while left<right:
            if only_letters[left]==only_letters[right]:
                left+=1
                right-=1
            else:
                return False
        return True

        

        
```
