```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        left=0
        right=len(numbers)-1
        s=0
        m=[]
        while left < right:
            s=numbers[left]+numbers[right]
            if s==target:
                break
            elif s<target:
                left+=1
            else:
                right-=1
        m.append(left+1)
        m.append(right+1)
        return m
        

        
        
```
