```python
class Solution:
    def intToRoman(self, num: int) -> str:
        s = ""
        
        # Düzgün sırada dictionary (böyükdən kiçiyə)
        a = {
            "M": 1000,
            "CM": 900,
            "D": 500,
            "CD": 400,
            "C": 100,
            "XC": 90,
            "L": 50,
            "XL": 40,
            "X": 10,
            "IX": 9,
            "V": 5,
            "IV": 4,
            "I": 1
        }

        # Key və value-ları birbaşa dictionary-dən istifadə edirik
        for k, v in a.items():
            while num >= v:
                s += k
                num -= v
        
        return s
```
