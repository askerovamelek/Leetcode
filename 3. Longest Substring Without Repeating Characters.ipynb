```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:

        if len(s) == 0:     # boş string üçün
            return 0

        m = s[0]
        best = 1
        j = 0

        for i in range(1, len(s)):
            if s[i] not in m:
                m += s[i]
                j = i
            else:
                while s[i] in m:
                    m = m[1:]
                m += s[i]
                j = i

            if len(m) > best:
                best = len(m)

        return best
```
