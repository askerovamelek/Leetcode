```python
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        a = "".join(str(i) for i in digits)  # hər rəqəmi string kimi birləşdiririk
        b = int(a) + 1                       # ədədə çevirib 1 əlavə edirik
        return [int(i) for i in str(b)]      # nəticəni yenidən int-lərə çeviririk
```
