```python
class Solution:
    def isHappy(self, n: int) -> bool:
        m = []
        
        while True:
            d = 0    # hər dəfə sıfırla

            while n > 0:
                p = n % 10
                n = n // 10
                d += p ** 2

            if d == 1:
                return True

            if d in m:    # eyni nəticə təkrar olsa loop var
                return False

            m.append(d)
            n = d
```
