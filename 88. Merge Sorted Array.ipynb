```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        merged_list = []
        for i in range(m):
            merged_list.append(nums1[i])
        for j in range(n):
            merged_list.append(nums2[j])

        merged_list.sort()

        # merged_list-ı nums1-ə geri yazırıq
        for k in range(m + n):
            nums1[k] = merged_list[k]
```
